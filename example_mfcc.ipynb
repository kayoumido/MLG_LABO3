```python
import numpy as np
from matplotlib import pyplot as pl
import os
import scipy.io.wavfile as wav
import statistics as stats

%matplotlib inline
```

```python
PATH = 'vowels/'
```

```python
sample_rate_1, X_1 = wav.read(os.path.join(PATH, 'namshaa.wav'))
sample_rate_2, X_2 = wav.read(os.path.join(PATH, 'nafshaa.wav'))
```

```python
pl.figure(figsize=(15,6))
pl.subplot(2,1,1)
pl.plot(X_1)
pl.xlim(0, len(X_1))
pl.title('X_1')
pl.grid()
pl.subplot(2,1,2)
pl.plot(X_2)
pl.xlim(0, len(X_2))
pl.title('X_2')
pl.grid()
pl.tight_layout()
```

```python
from python_speech_features import mfcc
```

```python
ceps_1 = mfcc(X_1, samplerate=sample_rate_1, nfft=1024)
ceps_2 = mfcc(X_2, samplerate=sample_rate_2, nfft=1024)
```
